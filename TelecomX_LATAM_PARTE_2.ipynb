```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2


datos_rank = datos.copy()


datos_rank["Churn"] = datos_rank["Churn"].map({"Yes": 1, "No": 0})


columnas_categoricas = datos_rank.select_dtypes(include=["object"]).columns


le = LabelEncoder()
for col in columnas_categoricas:
    datos_rank[col] = le.fit_transform(datos_rank[col])


X_cat = datos_rank[columnas_categoricas]
y = datos_rank["Churn"]


chi_scores, p_values = chi2(X_cat, y)


chi_datos = pd.DataFrame({
    "Variable": columnas_categoricas,
    "Chi2 Score": chi_scores,
    "p-value": p_values
})


chi_datos = chi_datos.sort_values(by="Chi2 Score", ascending=False)

chi_datos
```

# Análisis de Cancelación de Clientes (Churn) - TelecomX



# 1. Introducción

El objetivo de este análisis es **desarrollar modelos predictivos capaces de prever qué clientes tienen mayor probabilidad de cancelar sus servicios en la empresa TelecomX**. Para ello se implementaron **dos modelos de clasificación**.

Estos modelos permitieron **identificar patrones en los datos** y determinar qué variables influyen más en la decisión de un cliente de cancelar el servicio. Además, se evaluó el rendimiento de cada modelo utilizando **métricas de clasificación como la matriz de confusión y la curva ROC-AUC**.



# 2. Evaluación del rendimiento de los modelos

Se utilizaron **dos modelos diferentes** para analizar la cancelación de clientes.

## Regresión Logística

Este modelo permitió analizar la relación entre las variables del dataset y la **probabilidad de cancelación**.  

Su principal ventaja es la **interpretabilidad**, ya que los coeficientes del modelo muestran cómo cada variable influye en el churn.

## Random Forest

Este modelo basado en **árboles de decisión** mostró una mayor capacidad para **capturar relaciones complejas entre las variables**, permitiendo mejorar la precisión en la predicción de clientes que podrían cancelar el servicio.

Para evaluar los modelos se utilizaron las siguientes métricas:



### Resultados Matriz de Confusión

- **753 clientes** fueron correctamente clasificados como clientes que **no cancelan**.  
- **282 clientes** fueron correctamente identificados como clientes que **cancelan**.  
- **80 clientes** fueron clasificados incorrectamente como cancelación.  
- **294 clientes** que cancelaron **no fueron detectados por el modelo**.



### Accuracy

El modelo logra clasificar correctamente aproximadamente **74% de los clientes**.



### Recall

El **recall para churn (0.79)** indica que el modelo logra identificar **79% de los clientes que realmente cancelan**, lo cual es positivo para problemas de **retención de clientes**.



### Precision

Cuando el modelo predice que un cliente cancelará el servicio, **acierta aproximadamente el 51% de las veces**.



### F1-score

El **F1-score refleja el equilibrio entre precision y recall**, mostrando un desempeño **moderado del modelo para detectar cancelación**.


### Curva ROC y AUC

Esta curva muestra la relación entre:

- **Tasa de verdaderos positivos (TPR)**
- **Tasa de falsos positivos (FPR)**

Esto permite evaluar la capacidad del modelo para **distinguir entre clientes que cancelan y los que no**.

El análisis de la **curva ROC-AUC** permitió comparar el desempeño de ambos modelos y determinar cuál tiene una mejor capacidad para distinguir entre clientes que cancelan y los que permanecen activos.



# 3. Factores más relevantes en la cancelación de clientes

A partir del análisis de los **coeficientes de la Regresión Logística** y la **importancia de variables del modelo Random Forest**, se identificaron los factores que más influyen en la cancelación.

Entre las variables más relevantes se encuentran:



### Tipo de contrato

Los clientes con contratos **mes a mes (Month-to-Month)** presentan una mayor probabilidad de cancelar el servicio en comparación con aquellos que tienen **contratos a largo plazo**.



### Tiempo de permanencia (Tenure)

Los clientes con **menor tiempo en la empresa** tienden a cancelar con mayor frecuencia.  

Esto indica que **los primeros meses del cliente son críticos para su fidelización**.



### Cargos mensuales (Monthly Charges)

Clientes con **cargos mensuales más altos** muestran una mayor probabilidad de cancelación, posiblemente debido a la **percepción de alto costo del servicio**.



### Servicios adicionales

La presencia o ausencia de servicios adicionales como:

- soporte técnico  
- seguridad en línea  
- respaldo de datos  

también puede influir en la **permanencia del cliente**.

# 4. Estrategias de retención basadas en los resultados

A partir de los factores identificados, se pueden proponer diversas **estrategias para reducir la cancelación de clientes**.



### Programas de fidelización para nuevos clientes

Dado que los clientes con menor **tenure** tienen mayor probabilidad de cancelar, la empresa puede implementar **programas de fidelización durante los primeros meses del servicio**.



### Incentivar contratos a largo plazo

Ofrecer **descuentos o beneficios adicionales para contratos anuales o bianuales** puede reducir la probabilidad de cancelación en comparación con los contratos mensuales.


### Optimización de precios

Analizar la estructura de los **cargos mensuales** para identificar si existen servicios con **precios percibidos como altos** y ofrecer planes más competitivos.



### Promoción de servicios adicionales

Ofrecer paquetes que incluyan:

- soporte técnico  
- seguridad  
- beneficios adicionales  

puede aumentar el **valor percibido del servicio y mejorar la retención**.



# 5. Conclusión

El análisis realizado permitió identificar los **principales factores que influyen en la cancelación de clientes en TelecomX**.

Los modelos de **Regresión Logística y Random Forest** demostraron ser herramientas útiles para **predecir el churn y comprender las variables más relevantes en este fenómeno**.

Los resultados muestran que variables como:

- **tipo de contrato**
- **tiempo de permanencia**
- **cargos mensuales**

tienen un **impacto significativo en la decisión de los clientes de cancelar el servicio**.

Estos hallazgos permiten a la empresa implementar **estrategias de retención basadas en datos**, enfocadas en:

- mejorar la experiencia del cliente  
- optimizar los planes de servicio  
- fortalecer la fidelización  

lo cual puede contribuir a **reducir significativamente la tasa de cancelación**.
